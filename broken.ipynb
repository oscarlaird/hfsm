```python
import evadb
from tqdm import tqdm
cursor = evadb.connect(evadb_dir=("./eva_db")).cursor()
```

```python
# show pdfs
df = cursor.query("SELECT * FROM pdfs").df()
for row in tqdm(df.itertuples()):
    print(row.name, row.page, row.paragraph, row.data)
```

```python
# create pdfs table
cursor.execute("DROP TABLE IF EXISTS pdfs")
cursor.execute("CREATE TABLE pdfs (name TEXT, page INTEGER, paragraph INTEGER, data TEXT)")
```
